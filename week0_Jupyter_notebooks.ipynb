The use of Jupyter notebooks has been seen not only as a great educational tool, but also as a means for maximising research transparency and reproducibility, something that many research journals are [promoting these days](https://journals.plos.org/ploscompbiol/article?id=10.1371/journal.pcbi.1007007).

## First notebook

Now, you are ready to create a new Jupyter notebook, or just open one (as this one!). A Jupyter notebook is a single file (```file_name.ipynb```) that allows to collate nicely-written text with nicely-written and functional code. They consist of a collection of cells, each displaying either text or code. The story is simple:

1. To add a new cell, you just press the "plus" button next to save.


2. Each cell can be deemed as "Markdown" (i.e., text) or "Code", the last one by default. To change the cell type, just click in the cell and select the type from the dropdown menu.


3. If it is code, just write as you would do in Spyder or any other editor.


4. If it is Markdown, just write text, noting that most of LaTeX showrtcuts and notation work. Note that Markdown is minimalist, allowing few text edition just as **bold**, *italic*, and ~~crossed~~. Formulas are written in LaTeX, which means that they are very neat and flexible.


Of course, there are particularities for both code and text cells within Jupyter notebooks. Some will be described below. There are plenty of good resources available online (e.g., for [code](https://jupyter.org) and [Markdown](https://www.markdownguide.org/cheat-sheet/)).


**Example.** We want to compute and plot the following function

$$
f(n)=\sum_{i=0}^n 2^{-n} n,
$$

for $n$ integer in $[0,20]$. To do this, we create a new code cell, as we would do in Spyder:

````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Comment! For example, define arrays
n = np.arange(0., 20., 1)
f = np.cumsum(2**(-n)*n)

# Now, time to plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(n, f, 'o', label='f')
ax.set_xlabel('$n$', fontsize=13)
ax.legend(fontsize=13)
ax.set_title('The figure!', fontsize=15);
````

By adding a new cell, we can comment the result, and write whatever we want.

## Some code tips

Something that we may miss when writing code in Jupyter is the information that the code editor provides, just as the list of variables, or ask for help. Luckily, there exist workarounds! For example, we can check what the function ```cumsum``` from NumPy does just by

````python
help(np.cumsum)
````

Which gives us the official help from NumPy. Also, to check the variables created, we can simply call ```whos```:

````python
whos
````

Likewise, we can create functions, which, as all variables, will remain defined throughout the notebook. For example, we can define the square function equal to 1 between 0 and 1, and zero otherwise:

\begin{align}
\Pi (t) =
\begin{cases}
1 & \text{if } t\in[0,1]\\
0 & \text{otherwise}
\end{cases}.
\end{align}

````python
def square_fun(t):
    return (t>0) & (t<1)

# We now use it...
t = np.linspace(-1, 4, 100)
y = square_fun(t)

# ... and plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(t, y, '-')
ax.set_xlabel('$t$', fontsize=13)
ax.set_ylabel('$y$', fontsize=13)
ax.set_title('The second figure!', fontsize=15);
````

**Note:** In the previous case, we could have defined an inline function using the ```lambda``` command:

````python
y = lambda t: (t>0) & (t<1)

# ... and plot
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(t, y(t), '-')
ax.set_xlabel('$t$', fontsize=13)
ax.set_ylabel('$y$', fontsize=13)
ax.set_title('The second figure!', fontsize=15);
````

Simpler, right?
